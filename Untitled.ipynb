```python
from PIL import Image
from matplotlib import pyplot as plt
import pytesseract
img = Image.open("1.jpg")
area = (225, 450,
        1030, 600)
cropped_img = img.crop(area)
plt.imshow(cropped_img)
plt.title('my picture')
plt.show()
print(pytesseract.image_to_string(cropped_img))
```
